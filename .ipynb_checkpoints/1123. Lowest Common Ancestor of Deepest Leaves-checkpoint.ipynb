# 1123. Lowest Common Ancestor of Deepest Leaves
Given the `root` of a binary tree, return the lowest common ancestor of its deepest leaves.

Recall that:

The node of a binary tree is a leaf if and only if it has no children
The depth of the root of the tree is 0. if the depth of a node is d, the depth of each of its children is d + 1.
The lowest common ancestor of a set S of nodes, is the node A with the largest depth such that every node in S is in the subtree with root A.
 

## Example 1:


**Input:**
```python
root = [3,5,1,6,2,0,8,null,null,7,4]
```
**Output:** 
```python
[2,7,4]
```
**Explanation:** We return the node with value 2, colored in yellow in the diagram.
The nodes coloured in blue are the deepest leaf-nodes of the tree.
Note that nodes 6, 0, and 8 are also leaf nodes, but the depth of them is 2, but the depth of nodes 7 and 4 is 3.
## Example 2:

**Input:**
```python
root = [1]
```
**Output:**
```python
[1]
```
**Explanation:** The root is the deepest node in the tree, and it's the lca of itself.
## Example 3:

**Input:** 
```python
root = [0,1,3,null,2]
```
**Output:** 
```python
[2]
```
Explanation: The deepest leaf node in the tree is 2, the lca of one node is itself.
 

## Constraints:
```python
The number of nodes in the tree will be in the range [1, 1000].
0 <= Node.val <= 1000
The values of the nodes in the tree are unique.
```

Note: This question is the same as 865: https://leetcode.com/problems/smallest-subtree-with-all-the-deepest-nodes/

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
class Solution:
    def lcaDeepestLeaves(self, root):
        if not root:
            return None

        def dfs(node, depth):
            if not node.left and not node.right:
                return node, depth

            left, l_depth = dfs(node.left, depth+1) if node.left else (node, depth)
            right, r_depth = dfs(node.right, depth+1) if node.right else (node, depth)

            if l_depth == r_depth:
                return node, l_depth
            
            return (left, l_depth) if l_depth > r_depth else (right, r_depth)

        lca, _ = dfs(root, 0)
        return lca
````

````python
# Test case evaluator function
def run_tests():
    test_cases = [
        {
            'name': 'Test Case 1',
            'tree': TreeNode(1, TreeNode(2, TreeNode(4), TreeNode(5)), TreeNode(3, None, TreeNode(6, TreeNode(7), TreeNode(8)))),
            'expected': 6
        },
        {
            'name': 'Test Case 2',
            'tree': TreeNode(1),
            'expected': 1
        },
        {
            'name': 'Test Case 3',
            'tree': TreeNode(1, TreeNode(2), TreeNode(3)),
            'expected': 1
        },
        {
            'name': 'Test Case 4',
            'tree': TreeNode(1, TreeNode(2, TreeNode(3, TreeNode(4)))),
            'expected': 4
        },
        {
            'name': 'Test Case 5',
            'tree': TreeNode(1, TreeNode(2, TreeNode(4), TreeNode(5)), TreeNode(3, TreeNode(6), TreeNode(7))),
            'expected': 1
        },
        {
            'name': 'Test Case 6',
            'tree': TreeNode(1, TreeNode(2, None, TreeNode(3, TreeNode(4)))),
            'expected': 4
        },
    ]
    
    for case in test_cases:
        print(f"Running {case['name']}...")
        a = Solution()
        print(case['tree'])
        output = a.lcaDeepestLeaves(case['tree'])
        result = "PASS" if output and output.val == case['expected'] else f"FAIL (Expected: {case['expected']}, Got: {output.val if output else 'None'})"
        print(f"{case['name']} Result: {result}")
        print()
````

````python
# Run the tests
run_tests()
````
